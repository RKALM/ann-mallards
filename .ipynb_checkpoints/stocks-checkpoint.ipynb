# Importing the data

```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn import tree

tesla=pd.read_csv("TSLA.csv")
    
```

```python
print(tesla.head(5))
```

# Preparing the data

```python
df=pd.DataFrame(tesla)
close=df["Close"]
print(close)
```

```python
data=[]
counter=0

for x in range(0,len(close)):
    if counter==0:
        data.append("Up")
        counter +=1
    else:
        if close[x]>close[x-1]:
                data.append("Up")
        else:
              data.append("Down")
    
                
```

```python
df2=df
df2['Up/Down']=data
X=df2[['Open','Volume','Low','High','Close']]
y=df2["Up/Down"]
```

```python
print(X.head(5))
print(y.head(5))
```

# Creating the classification algorithm

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```

# Running and verifying the algorithm

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import cross_val_predict

y_train_pred = cross_val_predict(clf, X, y, cv=3)
print(accuracy_score(y, y_train_pred))
print(classification_report(y, y_train_pred))
print(confusion_matrix(y, y_train_pred))
```

# Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
clf2 = LogisticRegression()
```

```python
y_train_pred = cross_val_predict(clf2, X, y, cv=3)
print(accuracy_score(y, y_train_pred))
print(confusion_matrix(y, y_train_pred))
```

# Preparing and compiling Tensorflow and the algorithm

```python
#tf.keras.layers.experimental.preprocessing.Normalization
# normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer = tf.keras.layers.experimental.preprocessing.Normalization(axis=-1)
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

# Motivation about the performance
As I said in the previous lab, the stock value is either going up or down, and all the related data in the dataset follows the same characteristics, so it is difficult for a ML algorithm to make a prediction with better accuracy than 50%, and while simple algorithms like knn are stack in the coin-toss prediction, (50%-50%), a bit more advance algorithms like decision trees and deepl learning networks, can have a better accuracy that can reach up to 60% as they can detect some special characteristics that predict the next move of the stock price. In this lab the decision tree algorithm had the best prediction as it was favored by all the factors, (it can take sophisticated decisions with relatively small data), while the logistic regression freaked out as it had nothing to use for a prediction, (it is strange to a prediction from a binary algorithm about a binary output). At last, the deep learning network had the potential to do better predictions but that wasn't the case because of the limited params, and the relatively small amount of data, (it feels that it could use a bit more data).
